```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
nodes = ["A", "B", "C", "D", "E"]
G.add_nodes_from(nodes)
edges = [
    ("A", "B", 5),
    ("A", "C", 2),
    ("B", "C", 3),
    ("B", "D", 4),
    ("C", "D", 1),
    ("D", "E", 6),
    ("E", "A", 2)
]
G.add_weighted_edges_from(edges)
#it is just the degree of that node/n-1
degree_centrality = nx.degree_centrality(G)
#it is sum of all the weight connected to that node 
weighted_degree_centrality = {node: sum(weight for _, _, weight in G.edges(node, data="weight")) for node in G.nodes}
print("Degree Centrality:")
for node, centrality in degree_centrality.items():
    print(f"{node}: {centrality:.2f}")
print("\nWeighted Degree Centrality:")
for node, centrality in weighted_degree_centrality.items():
    print(f"{node}: {centrality}")
#ensures nodes to be fairly spaced
pos = nx.spring_layout(G)
#retrieves weight of all edges in the graph
edge_weights = nx.get_edge_attributes(G, 'weight')
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)
nx.draw_networkx_edges(G, pos, width=[weight for weight in edge_weights.values()])
plt.title("Graph with Weighted Edges")
plt.show()
```
