```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
nodes_component_1 = [1, 2, 3, 4]
edges_component_1 = [
    (1, 2),
    (2, 3),
    (3, 1),
    (1, 4),
    (4, 3),
]
G.add_nodes_from(nodes_component_1)
G.add_edges_from(edges_component_1)
nodes_component_2 = [5, 6, 7, 8]
edges_component_2 = [
    (5, 6),
    (6, 7),
    (7, 8),
    (8, 5),
    (5, 7),
]
G.add_nodes_from(nodes_component_2)
G.add_edges_from(edges_component_2)

# calculate connected components
connected_components = list(nx.connected_components(G))     #it will give list of connected components
component_sizes = [len(component) for component in connected_components]

print("Connected Components:")
for i, component in enumerate(connected_components):
    print(f"Component {i + 1}: {component}, Size: {component_sizes[i]}")

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)

# Assign colors to different components
color_map = []
colors = ["skyblue", "lightgreen"]
for node in G.nodes:
    for i, component in enumerate(connected_components):
        if node in component:
            color_map.append(colors[i])
            break

# Draw the graph
nx.draw_networkx_nodes(G, pos, node_size=700, node_color=color_map)
nx.draw_networkx_labels(G, pos, font_size=12, font_color="black")
nx.draw_networkx_edges(G, pos, edgelist=G.edges, edge_color="blue")

# Add a title
plt.title("Graph with Two Disconnected Components", fontsize=14)
plt.axis("off")
plt.show()
```
